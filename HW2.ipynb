Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
vector = [1,2,3,4,5]

def calc_len(vec):
    res = 0
    for i in vec:
        res+=i**2
    return math.sqrt(res)
    
calc_len(vector)
```

3. построить графики - окружность, эллипс, гипербола

```python
%matplotlib inline
x = []
y = []
y2 = []
p = 5
for i in range(1000):
    x_ = i/100
    x.append(x_)
    y.append(math.sqrt(2*p*x_))
    y2.append(-math.sqrt(2*p*x_))
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = []
x2 = []
y = []
y2 = []

for i in range(1500):
    r = 1000
    x.append(-i)
    x2.append(i)
    y.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))
    
plt.figure(figsize = (6, 6))    
plt.plot(x, y, color='r')
plt.plot(x, y2, color='r')
plt.plot(x2, y2, color='r')
plt.plot(x2, y, color='r')
plt.axis('scaled')
plt.show()
```

Докажите, что при ортогональном преобразовании сохраняется расстояние между точками


Пусть точки M1(x1, y1) и M2(x2, y2) посредством ортогонального преобразования переводятся соответствено в точки M'1 (x'1, y'1) и M'2(x'2, y'2). Требуется доказать, что отрезки M1M2 и M'1M'2 имеют равные длины.

Доказательство:

|M'1M'2|^2 = [x'2-x'1]^2 + [y'2-y'1]^2 + [a11(x2-x1) + a12(y2-y1)]^2 + [a21(x2-x1) + a22(y2-y1)]^2 =

= (a11^2 + a21^2) (x2-x1)^2 + (a12^2 + a22^2) (y2-y1)^2 + 2 (a11a12 + a21a22)(x2-x1)(y2-y1) =

= (x2-x1)^2 + (y2-y1)^2 = |M1M2|^2

Итак, |M1M2| = |M'1M'2|

3.1 Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_cart(r, phi):
    x = r * math.cos(phi)
    y = r * math.sin(phi)
    return x, y

polar_cart(1, 25)
```

4. Задание (в программе)
Решите систему уравнений:

y = x2 – 1
exp(x) + x(1 – y) = 1


y = x2 – 1
y = (exp(x) + x –  1) / x

```python
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.axis('scaled')
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
```

Решите систему уравнений и неравенств:

y = x2 – 1
exp(x) + x * (1 – y) > 1


y = x2 – 1
y = -1/x +2

```python
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)

plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, (- 1/x)+2 - y)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (0.5, 0.5))
x3, y3 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
